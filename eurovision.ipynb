```python
from pathlib import Path
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("eurovision.csv")
df.info()
```

```python
order = df['running_order']
rank = df['rank']
art = df['artist']
print(order[:3], rank[:3], art[:3])
```

```python
ress = pd.DataFrame({"order":order.values, 'values':rank.values, 'artist': art.values})
ress.head()
```

```python
plt.figure(dpi=100)
plt.rcParams['font.family'] = ['serif']
plt.scatter(order, rank, color='green', alpha=0.2) # меняем цвет, добавляем label
plt.title('') # перенос строки
plt.ylabel('rank')
plt.xlabel('order')
plt.show()
```

```python
plt.scatter(order[2000], rank[2000], c='green', label='group 1') # для Mathé Altéry
plt.title('') # перенос строки
plt.ylabel('rank')
plt.xlabel('order')
plt.show()
```
